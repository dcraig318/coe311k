```python
import math
import random
import numpy
import matplotlib.pyplot as plt

def approximatePi(number_of_shots): #Monte Carlo Pi approximation
    number_of_hits = 0
    for shot in range(number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            number_of_hits +=1
        
    return 4*number_of_hits/number_of_shots
```

```python
#Iterate from 10 to 10,000
shots = list(range(10,10001, 20))
errors = []

for n in shots: #Compute absolute error between approximated pi and true pi
    true_pi = numpy.pi
    calculated_pi = approximatePi(n)
    true_error = abs(calculated_pi - true_pi)
    errors.append(true_error)
```

```python
#Plot the results
plt.plot(shots, errors)
plt.xlabel("Number of Shots")
plt.ylabel("Absolute Error")
plt.show()
```
