```python
import math
import numpy as np
import matplotlib.pyplot as plt

#Taylor Series for Sine
def sine_taylor(x,n_terms):
    result = 0
    for n in range(n_terms):
        series = ((-1)**n) * (x**(2*n+1))/(math.factorial(2*n+1))
        result += series
    return result

#Approximate sin(x) at x=0 to x=10
x_values = np.arange(0,10,0.1)
error = {}
for n in [1,2,3,4]:
    approximation = np.array([sine_taylor(x,n) for x in x_values])
    error[n] = np.abs(approximation - np.sin(x_values))
    print(approximation)

#Plot errors
for n in [1,2,3,4]:
    plt.plot(x_values, error[n], label= f"{n} terms")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.legend()
plt.show()

#Discussion:
#As more terms are added, the value gets closer to the actual sin(x)
# so the graph shows the error decreasing as you add terms.
#Also, the error decreases significantly more for smaller x values,
# but they get larger as the x values increase.
# The Taylor series is more accurate for smaller x values, and needs more terms for the higher values of x
```
