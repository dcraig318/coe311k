## Part a

```python
import numpy as np
import time

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

#Jacobi
def jacobi(A,b,x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k+1
        x = x_new
    return x, max_iterations

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100000

t0 = time.perf_counter()
jacobi_solution, jacobi_iterations = jacobi(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

print(f"Jacobi Solution: {jacobi_solution}")
print(f"Iterations: {jacobi_iterations}")
print(f"Time: {t1 - t0}")

#Gauss-Seidel
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1
        
        x = x_new
        
    return x, max_iterations

t2 = time.perf_counter()
gauss_solution, gauss_iterations = gauss_seidel(A, b, x0, tol, max_iterations)
t3 = time.perf_counter()
print(f"\nGauss-Seidel Solution: {gauss_solution}")
print(f"Iterations: {gauss_iterations}")
print(f"Time: {t3 - t2}")
```

```python
A = np.array([[1, 	1], [1, 1.0001]])
b = np.array([1, 1])

t0 = time.perf_counter()
jacobi_solution, jacobi_iterations = jacobi(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

print(f"Jacobi Solution: {jacobi_solution}")
print(f"Iterations: {jacobi_iterations}")
print(f"Time: {t1 - t0}")

t2 = time.perf_counter()
gauss_solution, gauss_iterations = gauss_seidel(A, b, x0, tol, max_iterations)
t3 = time.perf_counter()
print(f"\nGauss-Seidel Solution: {gauss_solution}")
print(f"Iterations: {gauss_iterations}")
print(f"Time: {t3 - t2}")
```
