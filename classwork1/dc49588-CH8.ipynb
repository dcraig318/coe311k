```python
import numpy as np

def diag_dom(A):
    n = A.shape[0]
    for i in range(n):
        diag = abs(A[i,i])
        off_diag_sum = np.sum(np.abs(A[i,:])) - diag
        if diag < off_diag_sum:
            return False
    return True

A = np.array([[2,1],
            [1,1]])

print(A)
print(f"Is the matrix diagonally dominant? {diag_dom(A)}")
```
