```python
import numpy as np
import matplotlib.pyplot as plt

#Part 1

# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

#Plot
plt.scatter(X,Y)
plt.plot(X, m*X + c)
plt.show()

#Part 2

X = np.random.uniform(-1,1,5)
Y = np.random.uniform(-1,1,5)

idx = np.argsort(X)
X = X[idx]
Y = Y[idx]

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X,Y)
plt.plot(X, m*X + c)
plt.show()

#Part 3

X = np.arange(0, 1, 0.01)

m1 = 2
c1 = 6

perturbation = np.random.uniform(0, 1, len(X))
sign = np.random.choice([-1, 1], len(X))
Y = m1 * X + c1 + sign * perturbation

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Original equation: y = {m1}x + {c1} +/- random(0,1)")
print(f"Least squares fit: y = {m}x + {c}")

plt.scatter(X, Y)
plt.plot(X, m*X + c)
plt.show()
```
