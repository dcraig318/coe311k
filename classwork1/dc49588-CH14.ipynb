```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi  # Interval
N = 50  # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_forward = (f(x + h) - f(x)) / (h)

# Backward Difference Approximation
df_backward = (f(x) - f(x - h)) / (h)

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx, 'o--', label='Central Difference Approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x) using Central Difference")
plt.legend()
plt.show()
 
## Do 4 graphs on one plot comparing cos(x), Forward Difference, 
## Backwards Difference, and Central Difference Methods
plt.figure()
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_forward, 'o--', label='Forward Difference Approximation')
plt.plot(x, df_backward, 's--', label='Backward Difference Approximation')
plt.plot(x, df_dx, 'x--', label='Central Difference Approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Derivative Approximations of f(x) = sin(x)")
plt.legend()
plt.show()
```
