```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

sizes = [3 * (2 ** i) for i in range(8)]
naive_times = []
numpy_times = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    # Naive
    start = time.perf_counter()
    C_naive = naive_matrix_multiplication(A.tolist(), B.tolist())
    end = time.perf_counter()
    naive_times.append(end - start)

    # Numpy
    start = time.perf_counter()
    C_numpy = np.matmul(A, B)
    end = time.perf_counter()
    numpy_times.append(end - start)

print("Matrix sizes:", sizes)
print("Naive runtimes:", naive_times)
print("Numpy runtimes:", numpy_times)

matrix_numbers = list(range(1,9))

plt.plot(matrix_numbers, naive_times, label='Naive Multiplication')
plt.plot(matrix_numbers, numpy_times, label='Numpy Multiplication')
plt.grid(True)
plt.legend()
plt.show()
```
