```python
import math

def calc_hyp(a,b):
    h = math.sqrt(a*a + b*b)
    return h
```

```python
my_h = calc_hyp(3,4)
print(my_h)
```

```python
import math
import random

def approximatePi(number_of_shots):
    number_of_hits = 0
    for shot in range(number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            number_of_hits +=1
        
    return 4*number_of_hits/number_of_shots
```

```python
import random

value = random.uniform(-1,1)
print(value)
```
