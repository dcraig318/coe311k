```python
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

# Vary h, keep k=1
k = 1.0
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure(figsize=(10, 5))

for h in h_values:
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot only position y(x) for each h
    plt.plot(x_values, y_values, label=f'h = {h}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method for d²y/dx² = -ky with different h values (k = 1)")
plt.legend()
plt.grid(True)
plt.show()


# Vary k, keep h=0.01
h = 0.01
k_values = [1, 2, 5, 10]

plt.figure(figsize=(10, 5))

for k in k_values:
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    plt.plot(x_values, y_values, label=f'k = {k}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method for d²y/dx² = -ky with different k values (h = 0.01)")
plt.legend()
plt.grid(True)
plt.show()


# k = 10, h = 0.01
k = 10
h = 0.01
# Create arrays to store x, y, and v values
x_values = np.arange(x_start, x_end + h, h)
y_values = np.zeros(len(x_values))
v_values = np.zeros(len(x_values))

# Set initial conditions
y_values[0] = y_0
v_values[0] = v_0

# Euler's Method loop
for i in range(1, len(x_values)):
    x_n = x_values[i - 1]
    y_n = y_values[i - 1]
    v_n = v_values[i - 1]

    y_values[i] = y_n + h * v_n
    v_values[i] = v_n + h * (-k * y_n)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='Position y(x)')
plt.plot(x_values, v_values, label='Velocity v(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title("Solution of d²y/dx² = -10y using Euler's Method (h = 0.01)")
plt.legend()
plt.grid(True)
plt.show()
```

For different step sizes, the solution gets less accurate as h gets bigger. The oscillations grow in amplitude, showing the method becoming unstable. For small h, the solution is reasonable, but for larger h, after around h=0.2, the instability becomes obvious. With increased k, the oscillations became faster. The method got worse unless h stayed very small. For k=10 and h=0.01 the method works well, but if the h became larger it would not.
