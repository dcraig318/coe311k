```python
#In Class Challenge: Leibniz Series for Pi
import numpy

approx_pi = 0
for i in range(5000):
    series = (-1)**i / (2*i + 1)
    approx_pi += series
    if (i + 1)%100 == 0:
        calculated_pi = 4 * approx_pi
        error = abs((calculated_pi - numpy.pi)/numpy.pi)
        print(i+1, error)
```
